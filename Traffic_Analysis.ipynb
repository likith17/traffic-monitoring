```python
# Importing other libraries

import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline

```

```python


# Define paths and class names
val_img_dir = "/content/drive/MyDrive/final-project-code-submission-likith17/valid/images"
output_dir = "./predicted_images"
class_names = ['bus', 'car', 'truck', 'van']

# Create output directory
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process 10 validation images
img_files = [f for f in os.listdir(val_img_dir) if f.endswith(('.jpg', '.png'))][:10]
plt.figure(figsize=(15, 10))

for i, img_file in enumerate(img_files):
    img_path = os.path.join(val_img_dir, img_file)

    # Run inference
    results = pretrained_model(img_path, conf=0.25, iou=0.45)

    # Load image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Draw bounding boxes
    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = box.conf[0]
            cls = int(box.cls[0])
            # Handle class index
            if cls < len(class_names):
                label = f"{class_names[cls]} {conf:.2f}"
            else:
                label = f"Unknown (COCO cls {cls}) {conf:.2f}"


            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Save output
    output_path = os.path.join(output_dir, f"pred_{img_file}")
    cv2.imwrite(output_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

    # Display
    plt.subplot(2, 5, i + 1)
    plt.imshow(img)
    plt.title(img_file)
    plt.axis('off')

plt.tight_layout()
plt.show()
print(f"Saved predictions to {output_dir}")
```

```python
import os
from collections import Counter


base_path = "/content/drive/MyDrive/final-project-code-submission-likith17"

# Count images in each split
print("\nDataset Size:")
for split in ['train', 'valid', 'test']:
    img_dir = os.path.join(base_path, split, 'images')
    img_count = len([f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png'))])
    print(f"{split.capitalize()}: {img_count} images")

# Class distribution
class_names = ['bus', 'car', 'truck', 'van']

for split in ['train', 'valid']:
    lbl_dir = os.path.join(base_path, split, 'labels')
    class_counts = Counter()
    empty_files = 0

    for lbl_file in os.listdir(lbl_dir):
        if lbl_file.endswith('.txt'):
            with open(os.path.join(lbl_dir, lbl_file), 'r') as f:
                lines = f.readlines()
                if not lines or all(line.strip() == '' for line in lines):
                    empty_files += 1
                for line in lines:
                    if line.strip():
                        class_id = int(line.split()[0])
                        if 0 <= class_id < len(class_names):
                            class_counts[class_names[class_id]] += 1

    print(f"\n{split.capitalize()} Class Distribution:")
    for cls, count in class_counts.items():
        print(f"  {cls}: {count} objects")

    if empty_files:
        print(f"Warning: {empty_files} empty label files in {split}/labels")
```

```python
# Path to data.yaml
yaml_path = "/content/drive/MyDrive/final-project-code-submission-likith17/data.yaml"

#setting 'auto_balance = true' in the yaml file

with open(yaml_path, 'r') as file:
    lines = file.readlines()


if not any('auto_balance' in line for line in lines):
    lines.append("\nauto_balance: true\n")


with open(yaml_path, 'w') as file:
    file.writelines(lines)

print(" 'auto_balance: true' added to data.yaml.")
```

```python
# Fine-tuning the YOLOv11s on the data
from ultralytics import YOLO

# Load model
model = YOLO("/content/drive/MyDrive/final-project-code-submission-likith17/yolo11s.pt")

# Training configuration
results = model.train(
    data= "/content/drive/MyDrive/final-project-code-submission-likith17/data.y/content/aml",     # Path to the dataset configuration file
    epochs=100,              # Number of epochs to train for
    imgsz=416,               # Size of input images as integer
    device= 0 ,              # Device to run on
    patience=150,            # Epochs to wait for no observable improvement for early stopping of training
    batch=8,                 # Number of images per batch
    optimizer='auto',        # Optimizer to use
    lr0=0.0001,              # Initial learning rate
    lrf=0.1,                 # Final learning rate (lr0 * lrf)
    dropout=0.3,             # Automatically restore best model# Use dropout regularization
    freeze=15,               # freezing the first 15 layers to use its original weight
    seed=0                   # Random seed
)
```

```python
# Load fine-tuned model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Validate
metrics = model.val(data=yaml_path)

# Print metrics
print(f"\nValidation Summary (YOLOv11s):")
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP50-95: {metrics.box.map:.4f}")
print(f"Precision: {metrics.box.p.mean():.4f}")
print(f"Recall: {metrics.box.r.mean():.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract values from both models
metrics_names = ['Precision', 'Recall', 'mAP50', 'mAP50-95']
pretrained_values = [
    metrics_pre.box.p.mean(),
    metrics_pre.box.r.mean(),
    metrics_pre.box.map50,
    metrics_pre.box.map
]
finetuned_values = [
    metrics.box.p.mean(),
    metrics.box.r.mean(),
    metrics.box.map50,
    metrics.box.map
]

# Bar plot
x = np.arange(len(metrics_names))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, pretrained_values, width, label='Pretrained', color='lightcoral')
plt.bar(x + width/2, finetuned_values, width, label='Fine-Tuned', color='mediumseagreen')

# Annotate values
for i in range(len(metrics_names)):
    plt.text(x[i] - width/2, pretrained_values[i] + 0.01, f"{pretrained_values[i]:.2f}", ha='center', fontsize=9)
    plt.text(x[i] + width/2, finetuned_values[i] + 0.01, f"{finetuned_values[i]:.2f}", ha='center', fontsize=9)

plt.ylabel('Score')
plt.title('YOLOv11s Performance: Pretrained vs Fine-Tuned')
plt.xticks(x, metrics_names)
plt.ylim(0, 1.05)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python

# Process Video Frame-by-Frame with Overlay

vehicle_counts = []
sampled_frames = []  # Store a few frames to display later

cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

output_video_path = "traffic_output.mp4"
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

frame_idx = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, conf=0.4, iou=0.45)
    annotated_frame = results[0].plot()

    vehicle_count = len(results[0].boxes)
    vehicle_counts.append(vehicle_count)

    # Determine congestion level
    if vehicle_count < 5:
        congestion_level, color = "Low", (0, 255, 0)
    elif vehicle_count < 10:
        congestion_level, color = "Medium", (0, 255, 255)
    else:
        congestion_level, color = "High", (0, 0, 255)

    # Overlay text
    cv2.putText(annotated_frame, f"Vehicles: {vehicle_count}", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.putText(annotated_frame, f"Congestion: {congestion_level}", (20, 80),
                cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

    out.write(annotated_frame)

    # Save every 50th frame for display
    if frame_idx % 50 == 0:
        sampled_frames.append(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))

    frame_idx += 1

cap.release()
out.release()
print(f"Processed video saved: {output_video_path}")

```

```python

# Display Sampled Frames

plt.figure(figsize=(15, 8))
for i, f in enumerate(sampled_frames[:6]):  # Show some sampled frames
    plt.subplot(2, 3, i+1)
    plt.imshow(f)
    plt.axis('off')
    plt.title(f"Frame {i*50}")
plt.tight_layout()
plt.show()

```

```python

# Plot Vehicle Count Trend

plt.figure(figsize=(12, 5))
plt.plot(vehicle_counts, color='orange', linewidth=2)
plt.xlabel("Frame Number")
plt.ylabel("Vehicle Count")
plt.title("Vehicle Count Trend Over Video")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
